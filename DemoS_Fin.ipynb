```python
bounds=[(1,10),(10, 30),(80,100)]
popsize=20
dimensions = len(bounds)
pop = np.random.randint(0,10,(popsize, dimensions))
min_b, max_b = np.asarray(bounds).T
import math
OldRange = 10  
NewRange = (max_b-min_b)  
NewValue = (((pop - 0) * NewRange) / OldRange) + min_b
NewValue=NewValue.astype(int)
pop_denorm=NewValue
#diff = np.fabs(min_b - max_b)
#pop_denorm = min_b + pop * diff
pop_denorm
#pop
```

```python
import numpy
from sklearn import metrics   
l=list()
for ind in pop_denorm:
    clf=RandomForestClassifier(max_features=ind[0],max_leaf_nodes=ind[1],n_estimators=ind[2])
    clf.fit(X_train,Y_train)
    Y_pred=clf.predict(X_validation)
    Y_validation=Y_validation.astype(numpy.float32)
    Y_pred=Y_pred.astype(numpy.float32)
    
    print(np.sqrt(metrics.mean_squared_error(Y_validation, Y_pred)))
    l.append(np.sqrt(metrics.mean_squared_error(Y_validation, Y_pred)))
```

```python
l
fitness = np.asarray(l)
fitness
```

```python
best_idx = np.argmin(fitness)
best = pop_denorm[best_idx]
crossp=0.7
mut=0.8
for i in range(90):
    print(i)
    for j in range(popsize):
        index = [i for i in range(popsize) if i != j]
        a, b, c = pop[np.random.choice(index, 3, replace = False)]
        mutant = np.clip(a + mut * (b - c), 0, 9)
        cross_points = np.random.rand(dimensions) < crossp
        if not np.any(cross_points):
                cross_points[np.random.randint(0, dimensions)] = True
        trial = np.where(cross_points, mutant, pop[j])
        #trial_denorm = min_b + trial * diff
        NewValue = (((trial - 0) * NewRange) / OldRange) + min_b
        NewValue=NewValue.astype(int)
        trial_denorm=NewValue
        clf=RandomForestClassifier(max_features=trial_denorm[0],max_leaf_nodes=trial_denorm[1],n_estimators=trial_denorm[2])
        clf.fit(X_train,Y_train)
        Y_pred=clf.predict(X_validation)
        Y_validation=Y_validation.astype(numpy.float32)
        Y_pred=Y_pred.astype(numpy.float32)
        f=np.sqrt(metrics.mean_squared_error(Y_validation, Y_pred))
        if f < fitness[j]:
                fitness[j] = f
                pop[j] = trial
                if f < fitness[best_idx]:
                    best_idx = j
                    best = trial_denorm
       # yield best, fitness[best_idx]
        
        
```
